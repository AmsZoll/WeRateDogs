## Q1 : That is the most Dog poeple tweet about 

### The first question is
that occurred to my mind was what is the most liked dog, and I knew this by the number of tweets per dog number of favourates.I found out that Golden Retriever, Labrador Retriever, Pembroke, Chihuahua and Pug were the
most common breeds predicted in this dataset.

```python
df=pd.read_csv("twitter_archive_master.csv")
```

```python
df.groupby('dog_breed').count()['dog_score'].nlargest(5).plot(kind='bar')
plt.title("Top 5 Breeds of Dog Tweeted on WeRateDogs account",fontsize=16)
plt.ylabel("Dog Breed")
plt.xlabel("Number of Tweets");
df.dog_breed.value_counts()
```

### the second Question that is 
what is the relation between "Favorites and Retweets" and here the result is that The two variables appear to have similar distributions and positive correlation.

```python
Img = sns.regplot(x=df.retweet_count, y=df.favorites)
plt.title("Favorites and Retweets")
plt.xlabel('Retweets')
plt.ylabel('Favorites')
plt.show()
fig = Img.get_figure()
```

### the third Question that is 
The Relationship between timeStamp and favorites  and here the result is that month 1,6,7,12 has the most favorites count  that mean people were most active in this time. 

```python
df.timestamp = pd.to_datetime(df.timestamp, yearfirst = True)

df_re1=df['favorites'].groupby(df['timestamp'].dt.month).sum()

df_re1.plot(kind="bar",title='The Relationship between timeStamp and favorites ')
```

### The last question was 
what the first four dogs represented for all of the data and i found that those four dogs represent 25% of the total number of tweets in the dataset.
 

```python
labels = np.full(len(df.dog_breed.value_counts()), '', dtype = object)

labels[0] = 'Golden Retriever'
labels[1] = 'Labrador Retriever'
labels[2] = 'Pembroke'
labels[3] = 'Chihuahua'

df.dog_breed.value_counts().plot(kind = 'pie', labels = labels)
plt.title('Most 3 Common Dog Breeds')
```
